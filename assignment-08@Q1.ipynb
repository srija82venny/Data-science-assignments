# Data preprocessing

```python
df1=df.drop(['Type'],axis=1)
df1.head()
```

```python
df.info()
```

# Normalizing the data

```python
def norm_fucn(i):
    x = ((i-i.min())/(i.max()-i.min()))
    return(x)
df_norm = norm_fucn(df.iloc[:,1:])
df_norm
```

```python
std_df = StandardScaler().fit_transform(df1)
std_df.shape
```

```python
std_df
```

# Running PCA of standardized data.


```python
pca_std = PCA(random_state=14,n_components=3)
pca_std_df = pca_std.fit_transform(std_df)
pca_std_df
```

```python
#eigen values
pca_std.singular_values_
```

```python
# variance containing in each formed PCA
pca_std.explained_variance_ratio_*100
```

```python
# Cummulative variance ratio..

# this will give an idea of, at how many no. of PCAs, the cummulative addition of
#........variance will give much information..
cum_variance = np.cumsum(pca_std.explained_variance_ratio_*100)
cum_variance
```

# MODEL 1 - KMeans


```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 8))
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_) #criterion based on which K-means clustering works
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method - K-means with PCA Clustering')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters=3,init='k-means++',random_state=42)
y_kmeans = kmeans.fit_predict(df_norm)
y_kmeans
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_std_df[:,0],pca_std_df[:,1],cmap='tab10')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
pca_kmeans =pd.concat([df1.reset_index(drop = True), pd.DataFrame(df_norm)], axis= 1)
pca_kmeans.columns.values[-3: ] = ['Component 1', 'Component 2', 'Component 3']
pca_kmeans['Segment Kmeans PCA'] = y_kmeans
pca_kmeans
```

```python
pca_kmeans['Segment'] = pca_kmeans['Segment Kmeans PCA'].map({2:'First', 1:'Second', 0:'Third'})
```

```python
plt.figure(figsize= (10,8))
sns.scatterplot(x = pca_kmeans['Component 2'],y = pca_kmeans['Component 1'],hue=pca_kmeans['Segment'],palette=['g', 'r','c'])
plt.title('Clusters by PCA Componets')
plt.show()
```

```python
plt.figure(figsize= (10,8))
sns.scatterplot(x = pca_kmeans['Component 2'],y = pca_kmeans['Component 3'],hue=pca_kmeans['Segment'],palette=['g', 'r','c'])
plt.title('Clusters by PCA Componets')
plt.show()
```

```python
plt.figure(figsize= (10,8))
sns.scatterplot(x = pca_kmeans['Component 1'],y = pca_kmeans['Component 3'],hue=pca_kmeans['Segment'],palette=['g', 'r','c'])
plt.title('Clusters by PCA Componets')
plt.show()
```
